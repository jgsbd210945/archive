```python
# Jason Gordon
# PA6
# 30 November 2020

import random

def printcard(card): #Given. I put SUITS in this function because that's the only place it is used
    SUITS=u'\u2660'+u'\u2665'+u'\u2666'+u'\u2663'                            #The following are given:  |
    print('A23456789TJQK'[int(card)-1],SUITS[frac(card)-1],sep='',end=' ')                          #   |
                                                                                                    #   |
def frac(card):                                                                                     #   |
    return round(card %1 * 10)                                                                      #   |
                                                                                                    #   |
def rate(hand):  #sort and rate a poker hand                                                        #   |
    shand=sorted(hand)                                                                              #   |
    i,j,prod = 0,1,1                                                                                #   |
    while j < 5:                                                                                    #   |
        if int(shand[i]) == int(shand[j]):                                                          #   |
            prod=prod*2**(j-i)                                                                      #   |
            c=int(shand[j])  #what card made the 2,3,4 of a kind                                    #   |
        else:                                                                                       #   |
            i=j                                                                                     #   |
        j+=1                                                                                        #   |
    if prod == 1: #check for strait and flush and high strait                                       #   |
        d,f,e = 9,10,11                                                                             #   |
        for i in range(1,5):                                                                        #   |
            d = 1 if int(shand[i]) != int(shand[i-1])+1 else d                                      #   |
            e = 1 if int(shand[i]) != 8+i+int(shand[0]) else e                                      #   |
            f = 1 if frac(shand[0]) != frac(shand[i]) else f                                        #   |
        prod=d*e*f                                                                                  #   |
    prod = 9 if prod == 11 else prod # there aint no high strait                                    #   |
    prod = 1 if prod == 2 and 1 < c < 11 else prod #jacks or better                                 #   |
    return prod                                                                                     #   |

def draw(deck, hand): #This function replenishes the cards.
    for k in range (len(hand)): #I attempted to use for card in hand, but it unfortunately did not process correctly
        if hand[k] == 0:        #Instead I am using a for i in range() loop to change the list "hand" as a whole, which works.
            hand[k]=deck.pop(0)
    return hand
        
def printhand(hand): #Printing the hand. This function calls the printcard() function if there is something there and prints two spaces if there is no value there.
    for card in hand:
        if card == 0:
            print ("  ", end=" ")
        else:
            printcard(card)

def select(hand): #this function allows the user to select which cards to remove.
    printhand(hand)
    hold1=0 #Hold1-hold5 are used as placeholders if the user decides to recover a card.
    hold2=0 #I am using five of them because it makes it much easier to assign and re-assign cards without losing track.
    hold3=0
    hold4=0
    hold5=0
    e=int(input("Enter 1 through 5 to discard (or recover) a card or 0 to draw! ")) #Fenceposts. This is so that we can have a firm start to the function.
    while e != 0:
        if hand [e-1] == 0:
            if e == 1:
                hand[0]=hold1 #This is to recover the card.
            elif e == 2:      
                hand[1]=hold2
            elif e == 3:
                hand[2]=hold3
            elif e == 4:
                hand[3]=hold4
            elif e == 5:
                hand[4]=hold5
        else:
            if e == 1:
                hold1=hand[0] #The reason these two are separate is so that the card does not revert to 0 unless necessary.
            elif e == 2:      #If the "hand[e-1]=0" was put before these if/elif statements, the program could not save
                hold2=hand[1] #the cards and recovery would not work. If the line was put after the if/elif statements, the
            elif e == 3:      #if statement allowing it ("if hand[e-1]!=0") would always be active and even if a card was
                hold3=hand[2] #recovered through the if/elif statements, it would be reverted to 0, thus making the if/elif
            elif e == 4:      #statements useless.
                hold4=hand[3]
            elif e == 5:
                hold5=hand[4]
            hand[e-1]=0
        printhand(hand)
        e=int(input("Enter 1 through 5 to discard (or recover) a card or 0 to draw! "))
    return hand

def deal(deck): #This function deals the first five cards.
    hand=[]
    for i in range (0, 5): #To make this more concise, you could call draw twice with an empty hand the first time.
        hand.append(deck.pop(0))
    return hand

def newdeck(): #Makes a new deck and shuffles it
    deck=[]
    for i in range (1, 14):
        for j in range (1, 5):
            card=i+(j*0.1) #Since I needed an int, I decided against making this a string "i.j" because it would need to be made an int anyway.
            deck.append(card)
    random.shuffle(deck)
    return deck

def getbet(stake, prevbet): #Gets the bet at the beginning of each round.
    print("You have $", stake, ". How much would you like to bet? (0 to quit?) [Previous bet: $", prevbet, "] ", sep="")
    a=input() #I would have done input instead of print with the above, but I needed the "sep=""", which does not work in input functions.
    if a== "": #Hence, a blank input follows a print statement.
        a=prevbet
    elif int(a) < 0: #Prevents negative bets.
        print("You cannot bet a negative number. Please try again.")
        getbet(stake, prevbet)
    elif int(a) > stake: #Prevents bets larger than the amount of money they have.
        print("You do not have that much money. Please try again.")
        getbet(stake, prevbet) #The function is recursive, but a while loop would have also worked.
    elif int(a) == stake:
        print("Risky!")
    return int(a) #Since I need "a" to equal "" (for the bet being the previous bet), I decided not to make it an int at the
                  #input so that it could equal the string "".
def main(): #The control center of the function. It calls all other functions.
    hands={1: "Busted!", 2: "Jack or Better pair", 4: "Two Pair", 8: "Three of a Kind", 9: "Straight", 10: "Flush", 16: "Full House", 64: "Four of a Kind", 90: "Straight Flush", 110: "Royal Flush"}
    counts={1: -1, 2: 1, 4: 2, 8: 4, 9: 6, 10: 8, 16: 15, 64: 25, 90: 40, 110: 50} #I need these dictionaries to determine the value. They are here because they will only be used in main.
    balance=100 #Beginning balance
    prevbet=0 #Setting a value for prevbet
    while balance > 0: #This is the while loop that will repeat unless the balance is 0, when it will stop.
        a=getbet(balance, prevbet)
        prevbet=a
        if a==0:
            break #The while loop will also break if the user decides to leave, putting 0 as their bet.
        deck=newdeck()
        hand=deal(deck)
        newhand=select(hand)
        finalhand=draw(deck, newhand)
        printhand(finalhand)
        result=rate(finalhand) #The above sequence calls all the functions in order in a round.
        wordresult=hands[result] #To determine the value, I could call a different function, but I would need to use variables in the main function anyway.
        print("Your hand is:", wordresult) #Also, the assignment does not ask for this function.
        numresult=counts[result]*a #Instead, I search the dictionaries given for the correct results.
        if numresult < 0:
            print("You lose:", numresult*-1) #If the user busted, the appropriate message is given.
        else:
            print("You win:", numresult) #The dictionaries also provide the results needed.
        balance+=numresult #And the result is added to the total. If it is negative, it will add the negative, making the balance smaller.
    print("Thank you for playing! You leave with $", balance, sep="") #This is outside of the while loop, and it is the goodbye message.
    
main()
```
