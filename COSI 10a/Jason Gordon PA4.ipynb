```python
#Jason Gordon
#PA4
#28 October 2020

import random #40 characters/line

def processpara(x1): #this function is named as such because it processes a paragraph and returns it as a list of words.
    x=""
    for j in range (len(x1)): #x1 enters as a list. We need to convert that to a string to split it again.
        x+=(x1[j])
    y1=x.split("\n") #We repeat the process for the remaining types of whitespace.
    y=""
    for j in range (len(y1)):
        for i in range(len(y1[j])):
            y+=(y1[j][i])
        y+="\t"
    z1=y.split("\t")
    z=""
    for j in range (len(z1)):
        for i in range(len(z1[j])):
            z+=(z1[j][i])
        z+=" "
    a=z.split(" ") #since z.split returns a list, we do not need to convert it to a string because we want a list anyway.
    return a

def printpara(words, length): #this is the workhorse of the function. It calls other functions and is the sort-of "nexus" of the function.
    while len(words) != 0:
        line=getline(words, length)
        if len(words) == 0: #this is to find the end of the paragraph. The function will not work if there is one word and it will look odd if there are not many words.
            spaces=[1]*(len(line)-1) #Thus, we add a place where the spaces are just one.
            spaces+=[0]
            printline(line, spaces)
        else:
            x=addspace(line, length)
            printline(line, x)
        print("\n", end="")
            
def addspace (line, length): #This figures out the in-between spaces for the function.
    total=0
    spaces=[1]*(len(line)-1) #This is to create a list the exact amount we need for the line.
    spaces+=[0] #This is to create a blank space on the end so as to not have extra spaces.
    for i in range (len(line)):
        total+=len(line[i])
    total+=len(spaces)
    while total<length: #This while loop 
        y=random.randint(0, (len(line)-2)) #This is to ensure the last value will be 0.
        spaces[y]+=1
        total+=1
    return spaces
    
def printline(line, spaces): #This function prints the lines. I found it best to use two separate lists in order to correlate the right amount of spaces to the word.
    for i in range (len(line)):
        print(line[i], end="")
        print(" "*spaces[i], end="")
        print(" ", end="")
    
def getline(words, length): #This function was provided.
    ans=[]
    total=0
    while (length>total) and 0 != len(words):
        word=words.pop(0)
        total += len(word)+1
        ans.append(word)
    if total > length:
        words.insert(0,ans.pop())
    return ans

def main(file):
    x1=file.split("\n\n") #Split along paragraphs.
    x2=[]
    for i in range (len(x1)):
        x2+=[processpara(x1[i])] #This helps keep the main function short. We should convert it to a list in order to avoid having the next function split by character instead of list item.
    for j in range (len(x2)):
        g=printpara(x2[j], 40) #This sends it to processing the paragraph. The length is 40 because we're looking to get it to that line length.
        print() #this prints an extra space in between the paragraphs.



main("I am typing a bit of nonsense in order to test my text justification program. The goal is to split a file into paragraphs using a unique separator,\nand then for each paragraph, split it into lines and add spaces so the each line prints the same length.\n\nWe are ignoring punctuation and carriage returns.\nThe only caveat is that if the last line only has one word in it, it should be left justified.\n\nThe goal of the program is to be a primitive simulation of one of the functions of a word processor, the most ubiquitous of software applications which used to have a vibrant ecosystem of variants, but which has collapsed into a monopoly governed by Microsoft word.\nOther parts of word processing require paging files into buffers, spell checking, display management and keyboard processing.")
```
