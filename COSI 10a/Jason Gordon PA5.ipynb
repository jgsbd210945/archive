```python
#Jason Gordon
#PA5
#9 November 2020

import random
import itertools #We need to import these to run our functions

def printboard(clue, guesses, answers): #These functions are put in reverse order from what is called first to what is
    for m in range (len(clue)):         #called last. This is so that functions that call other functions already have
        print(clue[m], end="")          #the function defined.
    print()
    for word in answers:
        if word in guesses:  #This function prints the board. If the word is in the set "guesses", then it will print the
            print(word)      #word. If it isn't, it will print a series of blanks followed by spaces to show the length of
        if word not in guesses: #the word. It will also print the clue.
            print("_ "*(len(word)))
    

def twist(clue, answers): #This function is what the player sees, and it will be the workhorse after the clue and answers
    random.shuffle(clue)  #are selected. First, it shuffles the clue (otherwise, it will be just the word), gives a value
    guesses=set()         #to guesses, and calls printboard for the first time. Then, it calls the input.
    printboard(clue, guesses, answers) #We need to do this because we want the board and the clue before they guess, and they
    d=str(input("Guess a word, or q to quit, t to twist, or h for a hint: ")) #need to guess before the while loop is initiated.
    while d != "q":
        if d == "h": #Extra Credit. It will reveal a random word that has not been revealed already.
            if len(guesses)!=len(answers): #To safeguard infinite loops if all the answers have been guessed.
                pick=random.randint(1, len(answers))-1
                while answers[pick] in guesses: #Will ensure that there will always be a new word revealed.
                    pick=random.randint(1, len(answers))-1
                guesses.add(answers[pick])
        elif d == "t":
            random.shuffle(clue) #This is to re-shuffle the word. It will simply change the order of the letters.
        elif d in answers: #These use elif functions to possibly save a tiny bit of processing power, as the kernel can
            guesses.add(d) #skip over these if there is a previous contingency
        printboard(clue, guesses, answers) #No matter what, it will print the board. If the guess is in the list answers, it
        if len(guesses)==len(answers):     #will add it to guesses.
            print("You win! Press \'q\' to quit.") #So that it specifically notifies them that they win
        d=str(input("Guess a word, or q to quit, t to twist, or h for a hint: ")) #For the next guess.
    print("Thanks for playing!") #Final message. Will print after they quit.

def getwords(clue, n, WORDS): #We need WORDS here, or else we will not be able to confirm that the words are in the list.
    z1=itertools.permutations(clue, n) #This returns every iteration (in the form of a tuple) of length n of these letters.
    z=list(z1) #This is to convert them into a list so that it can be processed.
    a1=[]
    for j in range (len(z)):
        a1.append("".join(z[j])) #a1 will make a list of these, but it will form the string first. This will take it out of
    a=set()                      #tuple form. We then convert it to a set to remove any duplicates as well as to sort.
    for k in range (len(a1)):
        if a1[k] in WORDS: #Before adding it into the set, we may as well ensure that it is in the list.
            a.add(a1[k])
    return sorted(a) #This will sort the set and return it to getallwords..

def getallwords(clue, WORDS): #This is so we can move it to getwords.
    b=[]
    for n in range (3, 8):
        c=getwords(clue, n, WORDS) #c gets the sorted set for all n, and it will redefine itself as the for loop works through.
        for l in range (len(c)):
            b.append(c[l]) #This is to move them to a list, as well as add the set before it redefines itself.
    return b #returns to main

def main(x):
    with open(x) as file: #opens the file and starts it at the beginning.
        y=file.read()
        WORDS=y.split() #this is to get a list of words. We will be using this to confirm the list of words.
        select=[]
        for i in range (len(WORDS)):
            if len(WORDS[i]) == 7: #This only chooses 7 letter words as the max. It can be set to any length.
                select+=[WORDS[i]]
        choice=random.choice(select) #This randomly chooses one of those 7 letter words, but we would like it as a list of letters.
        clue=list(choice) #Hence, we use list(choice) to break apart the string into letters.
        answers=getallwords(clue, WORDS) #calls getallwords and gets the set of answers
        twist(clue, answers) #This runs the game.

main("words7.txt")
```
